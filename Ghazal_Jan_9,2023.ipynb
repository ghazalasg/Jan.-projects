```python
# creating the first model


home_data.columns
```

```python
Y=home_data.median_house_value
print(Y)
```

```python
feature_names=['housing_median_age', 'total_rooms',
       'total_bedrooms']
X=home_data[feature_names]


X.describe()
X.head()
```

```python
# model validation
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
X_tr,X_val,Y_tr,Y_val= train_test_split(X,Y, random_state=1) #here i should respect the orders of variables X_tr, X_val are chosen from X and
#then Y_tr and Y_val are chosen from Y

X_tr.shape
print(X_tr)
```

```python
Y_tr.shape
print(Y_tr)
Y_tr.isnull()
```

```python
from sklearn.tree import DecisionTreeRegressor
california_housing_model=DecisionTreeRegressor(random_state=1, max_depth=20) # fort max_depth=20 the absolute error value is less than 101000, while it
# was around 101000 before changing max_depth
california_housing_model.fit(X_tr,Y_tr)
prediction=california_housing_model.predict(X_val)
print(prediction)
```

```python
# befor intoducing model selection i used this code to evaluate the model accuracy

#prediction=california_housing_model.predict(X.head())
#print(prediction)
```

```python
model_error=mean_absolute_error(Y_val, prediction)
print(model_error)
```
